# This tutorial is for LinkedList:
**In this tutorial we will implement the following functions:**

*Insert_at_beginning*: This function will add the input value at the beginning of the linked list.

*add_at_end*: This function is for adding the value at the end of the linked list.

*InsertValues*: This is for adding a list of values in the linked list.


*Get_length*: Get the length of the linked list.

*Remove_at_index*: This is for removing the values at the provided index.

*Remove_by_value*: The function for removing the node based n the provided value.

*Insert_at_index*: The function to insert the value at the specified index.

*Insert_next_value_by_given_values*: For inserting the values after provided value without basing on the index.

*PrintList*: This is for displaying the status of the Linked List.


```python
class Node:
  def __init__(self,data=None,place=None):
    self.stored_value= data
    self.next_node= place


class LinkedList:

  def __init__(self):

    self.head_index=None

  def Insert_at_beginning(self,data):
    node=Node(data,self.head_index)
    self.head_index=node

  def add_at_end(self,data):

    if self.head_index is None:
      self.head_index=Node(data,None)
      return

    iterator=self.head_index

    while iterator.next_node:
      iterator=iterator.next_node
    iterator.next_node=Node(data,None)
  def InsertValue(self,data_list):
    self.head_index=None

    for data in data_list:
      self.add_at_end(data)

  def GetLength(self):
    count=0
    iterator=self.head_index
    while iterator:
      count+=1
      iterator=iterator.next_node
    return count

  def Remove_at_index(self,index):
    if self.head_index is None:
      raise Exception('Your linkedlist is empty.')



    if index < 0 and index> self.GetLength()-1:
      raise Exception('Your index is out list size')


    if index==0:
      self.head_index=self.head_index.next_node
      return


    count=0
    iterator=self.head_index

    while iterator:
      if count==index-1:

        iterator.next_node=iterator.next_node.next_node
        break
      iterator =iterator.next_node
      count+=1
    return

  def Remove_by_value(self,data):
    if self.head_index is None:
      print('The linked list is empty.')
      return
    if self.head_index.stored_value==data:
      self.head_index=self.head_index.next_node
      print('It was simple since data was at the head.')
      return

    count = 0
    iterator = self.head_index

    while iterator:
      if iterator.stored_value==data:
        self.Remove_at_index(count)
        print('\n Data was at %sth index'%count)
        break
      iterator=iterator.next_node
      count+=1

  def Insert_at_Index(self,index,data):

    if self.head_index is None:
      raise Exception('Your linkedlist is empty.')



    if index < 0 and index> self.GetLength()-1:
      raise Exception('Your index is out list size')

    if index==0:
      self.Insert_at_beginning(data)

      return
    iterator=self.head_index

    count=0

    while iterator:
      if index-1 == count:
        new_node=Node(data,iterator.next_node)
        iterator.next_node=new_node

        break
      iterator= iterator.next_node
      count+=1

  def Insert_after_value(self,index,data):
    if index==0:
      self.Insert_at_Index(1,data)
      return
    if index== self.GetLength()-1:
      self.add_at_end(data)
      return

    count=0
    iterator=self.head_index

    while iterator:
      if count == self.GetLength()-1:
        self.add_at_end(data)
        break
      if count==index and index+1< self.GetLength():
        self.Insert_at_Index(index+1,data)
      iterator=iterator.next_node
      count +=1

  def Insert_next_value_by_given_values(self,prev_val,after_val):
    # print('The entered values are : ',prev_val,after_val)
    if self.head_index is None:
      print('Your linked list is None.')
      return

    if self.head_index.stored_value==prev_val:
      self.Insert_after_value(0,after_val)

      return

    count=0

    iterator=self.head_index

    while iterator:
      # print(count)
      if count==self.GetLength()-1 and iterator.stored_value != prev_val:
        print('The data you provided for check are not in your linked list.')
        break
      elif iterator.stored_value==prev_val:
        self.Insert_after_value(count,after_val)
        print('Data is inserted at index : ',count+1)
        break
      iterator=iterator.next_node
      count+=1



  def PrintList(self):

    if self.head_index is None:
      print('The linkedlist is empty.')
      return

    iterator=self.head_index
    values=''
    while iterator:

      values+=str(iterator.stored_value)+'----->'
      iterator=iterator.next_node
    print(values)
```
